```python
import pandas as pd

# Sample dataset
data = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'product_A_sales': [150, 200, 250, 300, 400, 500, 450],
    'product_B_sales': [100, 120, 190, 220, 260, 300, 320],
    'ad_spend': [30, 35, 50, 60, 70, 90, 85]
})

print(data)
```

```python
sns.scatterplot(data=data, x='ad_spend', y='product_A_sales')
```

```python
sns.lineplot(data=data, x='day', y='product_A_sales', label='Product A')
sns.lineplot(data=data, x='day', y='product_B_sales', label='Product B')
plt.title('Sales over the week')
plt.ylabel('Units sold')
plt.show()
```

```python
avg_sales = {
    'Product A': data['product_A_sales'].mean(),
    'Product B': data['product_B_sales'].mean()
}

avg_df = pd.DataFrame(list(avg_sales.items()), columns=['Product', 'Avg_Sales'])

sns.barplot(data=avg_df, x='Product', y='Avg_Sales')
plt.title("Average Weekly Sales by Product")
plt.show()
```
